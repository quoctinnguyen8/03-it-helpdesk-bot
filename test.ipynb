```python
import gradio as gr


def greet(name):
    return f"Hello, {name}!"


with gr.Blocks() as demo:
    gr.Markdown("# Greeting App")
    name_input = gr.Textbox(label="Enter your name")
    greet_button = gr.Button("Greet")
    output = gr.Textbox(label="Greeting")

    greet_button.click(fn=greet, inputs=[name_input], outputs=[output])

demo.launch()
```

```python
import pandas as pd


def read_csv(file):
    if file is None:
        return "No available file"

    df = pd.read_csv(file.name)
    return df.head()


with gr.Blocks() as demo:
    gr.Markdown("# CSV Reader")
    file_input = gr.File(label="Upload CSV File", file_types=[".csv"])
    table = gr.Dataframe(label="CSV Content")

    file_input.change(fn=read_csv, inputs=[file_input], outputs=[table])

demo.launch()
```

```python
import gradio as gr


def show_choice(choice):
    return f"You selected: {choice}"


with gr.Blocks() as interface:
    gr.Markdown("# Choose an Option")
    dropdown = gr.Dropdown(
        label="Select an option",
        choices=["Option 1", "Option 2", "Option 3"],
        interactive=True
    )

    output = gr.Textbox()

    dropdown.change(fn=show_choice, inputs=[dropdown], outputs=[output])

interface.launch()
```

```python
import gradio as gr


def play_audio(file):
    if file is None:
        return "No audio file uploaded"
    return "Playing audio..."


with gr.Blocks() as interface:
    gr.Markdown("# Audio Player")
    audio_input = gr.Audio(label="Upload Audio", interactive=True)
    output = gr.Textbox()


interface.launch()
```

```python
import gradio as gr


def echo_audio(audio):
    file_path, _ = audio
    return file_path


with gr.Blocks() as demo:
    gr.Markdown("Ghi am va nghe lai")
    mic_input = gr.Audio(label="Ghi am tai day", sources=[
                         'microphone'], interactive=True)

    playback = gr.Audio(label="Am thanh da ghi", interactive=False)
    mic_input.change(fn=echo_audio, inputs=[mic_input], outputs=[playback])

demo.launch()
```
